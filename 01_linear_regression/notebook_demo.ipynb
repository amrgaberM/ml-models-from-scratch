```python
from linear_regression import linearRegressionScratch
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Simple linear dataset: y = 2x + 1
X = np.array([1, 2, 3, 4, 5])
y = np.array([3, 5, 7, 9, 11])
```

```python
model=linearRegressionScratch(Learning_Rate=0.01,n_iters=1000,verbose=True)
model.fit(X,y)
```

```python
y_pred = model.predict(X)

print("Predictions:", y_pred)
print(f"Final weights: w = {model.w}, b = {model.b:.4f}")
```

```python
plt.figure(figsize=(8, 5))
plt.scatter(X, y, label='Actual Data', color='blue')
plt.plot(X, y_pred, color='red', label='Model Prediction')
plt.title("Linear Regression Fit")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Bonus: add random noise to y
np.random.seed(42)
y_noisy = y + np.random.randn(*y.shape)

model2 = linearRegressionScratch(Learning_Rate=0.01, n_iters=1000)
model2.fit(X, y_noisy)
y_pred_noisy = model2.predict(X)

plt.figure(figsize=(8, 5))
plt.scatter(X, y_noisy, label='Noisy Data', color='orange')
plt.plot(X, y_pred_noisy, color='green', label='Model Fit on Noisy Data')
plt.title("Linear Regression with Noisy Data")
plt.xlabel("X")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```
